```python
import numpy as np
import matplotlib.pyplot as plt
```

# define my function: $f(x) = 2^x$

```python
def myFunction(x):
    f = 2**x
    return f
```

# define the derivative of my function: $f'(x) = 2^x\ln 2$

```python
def myDerivativeFunction(x):
    Df = 2**x * np.log(2)
    return Df
```

# first-order Taylor approximation : $f(x) \approx f(a)+f'(a)(x-a)$

```python
def TaylorApproximation(x,a):
    TA = 2**a + 2**a * np.log(2)*(2**x-a)
    return TA
```

# define the domain of the function: $x = [-5 : 0.1 : 5]$

```python
x = np.arange(-5, 5, 0.1)
```

# compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

# pick 3 points in the domain: $a=(-1, 0 , 1)$

```python
TA_n1 =TaylorApproximation(x,-1)
TA_0 = TaylorApproximation(x,0)
TA_1 = TaylorApproximation(x,1)
```

# plot the graphs for the function, its derivative and Taylor approximation

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.plot(x, TA_n1, 'g', label="Taylor(a= -1)")
plt.plot(x, TA_0, 'y', label="Taylor(a= 0)")
plt.plot(x, TA_1, 'k', label="Taylor(a= 1)")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
