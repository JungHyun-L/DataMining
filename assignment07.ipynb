```python
import numpy as np
import matplotlib.pyplot as plt
```

# define least square approximation function: $ \hat{x}=(A^{\rm T}A)^{-1}A^{\rm T}b$

```python
def leastsquareApprox(xMat,yMat) :
    xTx= np.dot(xMat.T,xMat)
    middle=np.dot(np.linalg.inv(xTx),xMat.T)
    ws=np.dot(middle,yMat)
    return (ws)
```

# define the function automatically generating  $A$ according to degree

```python
def genMat(degree,x) :
    if degree==0:
        print("차수는 1 이상이여야 합니다")
    
    oneVec=np.ones(x.size)
    oneVec=np.reshape(oneVec,(x.size,1))
    xMat= oneVec
    
    for i in range(degree):
        result=np.reshape(x**(i+1),(x.size,1))
        xMat=np.hstack((xMat,result))
    
    return(xMat)
```

# plot the noisy and clean data

```python
num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'r.', x, y2, 'k.')
plt.show()
```

# degree 3 : polynomial curves and residual

```python
xMat=genMat(3,x)
yMat=np.reshape(y1,(y1.size,1))

ws=leastsquareApprox(xMat,yMat)
yHat=np.dot(xMat,ws)

residual= (y1-yHat.T)**2

plt.subplot(2, 1, 1)
plt.plot(x, y1, 'b.', x, yHat, 'r-')
plt.title('degree=3')
plt.ylabel('polynomial curve')


plt.subplot(2, 1, 2)
plt.ylabel('residual')
plt.axis([-11, 11, 0, 100])
plt.plot(x, residual.flatten(), 'b.')
```
